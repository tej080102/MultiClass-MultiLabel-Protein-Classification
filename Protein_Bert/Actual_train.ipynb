```python
from transformers import BertModel, BertTokenizer
import re
tokenizer = BertTokenizer.from_pretrained("Rostlab/prot_bert_bfd_localization", do_lower_case=False )
model = BertModel.from_pretrained("Rostlab/prot_bert_bfd_localization").to('cuda')
```

```python
model
```

```python
for index, row in df.iterrows():
    x=row['Sequence']
    sequence=' '.join(char for char in x)
    df.at[index,'Sequence']=sequence
```

```python
import torch
import torch.nn as nn
import torch.cuda.amp as amp
from torch.nn.parallel import DataParallel

# Assuming you have defined 'model' and 'tokenizer' elsewhere

# Define your feature extraction function
def features(sequence):
    # sequence = re.sub(r"[UZOB]", "X", sequence)
    # sequence=' '.join(char for char in sequence)
    # print(sequence)
    with amp.autocast(), torch.no_grad():
        encoded_input = tokenizer(sequence, return_tensors='pt').to('cuda')

        # Wrap your model with DataParallel
        model_parallel = DataParallel(model)

        output = model_parallel(**encoded_input)

        c = output.pooler_output.detach().cpu().numpy().flatten()
        del encoded_input
        del output
        torch.cuda.empty_cache()
        return list(c)


```
