```python
def parse_features(feature_string):
    feature_string = str(feature_string)
    feature_list = feature_string.strip('[]').split(',')
    feature_list = [float(val) for val in feature_list]
    return feature_list
```

```python
trainMM['feature_list'] = trainMM['feature'].apply(parse_features)# Create a new DataFrame with each feature as a separate column
trainMM = pd.DataFrame(trainMM['feature_list'].tolist())
trainMM.columns = [f'Feat{i}' for i in range(1, 1025)]
```

```python
testMM['feature_list'] = testMM['feature'].apply(parse_features)# Create a new DataFrame with each feature as a separate column
testMM = pd.DataFrame(testMM['feature_list'].tolist())
testMM.columns = [f'Feat{i}' for i in range(1, 1025)]
```

```python
col=[]
for m in Y_train.columns:
    col.append(m)
```

```python
trainMM.fillna(0,inplace=True)
testMM.fillna(0,inplace=True)
```

```python
params = {
    'max_depth': 100,
    'min_samples_split': 10,
    'min_samples_leaf': 5,
    'max_features': 'sqrt',
    'criterion': 'gini',
    'class_weight': 'balanced'
}
models=[]
from sklearn.tree import DecisionTreeClassifier
for i in range(len(col)):
    knn=DecisionTreeClassifier()
    knn.fit(trainMM, Y_train[col[i]]) 
    # print(models.predict(X2))
    models.append(knn)
```

```python
for i in range(len(models)):
    print(models[i].score(trainMM, Y_train[col[i]]))
```

```python
dd1=[]
for i in range(len(models)):
    dd1.append(models[i].predict(trainMM))
```
