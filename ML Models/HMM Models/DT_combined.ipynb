```python
trainMM.dropna(inplace=True)
testMM.dropna(inplace=True)
trainPse.dropna(inplace=True)
testPse.dropna(inplace=True)
```

```python
C1=[]
for i in range(1,401):
    C1.append("Feat_"+str(i))
```

```python
XTrainMM=trainMM[C1]
XTestMM=testMM[C1]
```

```python
C2=[]
for i in range(1,321):
    C2.append("Feat_"+str(i))
```

```python
XTrainPse=trainPse[C2]
XTestPse=testPse[C2]
```

```python
XTrainPse.columns=[i for i in range(320)]
XTestPse.columns=[i for i in range(320)]
```

```python
XTrainMM.columns=[i for i in range(320,720)]
XTestMM.columns=[i for i in range(320,720)]
```
